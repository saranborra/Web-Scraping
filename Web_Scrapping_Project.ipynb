```python
import requests #urllib,selenium
import urllib
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url ='https://www.cars24.com/buy-used-cars-hyderabad/?itm_medium=sticky_header&itm_source=Cars24Website&page=1'
```

```python
url_list =[]

for i in range(0,27):
    data = requests.get('https://www.cars24.com/buy-used-cars-hyderabad/?itm_medium=sticky_header&itm_source=Cars24Website&page=1',str(+i))
    url_list.append(data)
```

```python
data
```

```python
car_names =[]
model_year=[]
fuel_type=[]
Distance_travelled=[]
price =[]
Owner_type=[]
Transmission=[]
Rto=[]
Insurance=[]
view_details_url_list=[]
for i in url_list:
    soup = BeautifulSoup(i.text)
    
    data_tag_car_names = soup.find_all('h3')
    for j in data_tag_car_names:
        car_names.append(j.text)
        
    data_tag_model_year = soup.find_all('li',itemprop='modelDate')
    for k in data_tag_model_year:
        model_year.append(k.text)
        
    data_tag_fuel_type = soup.find_all('span',class_='_1BHEb')
    for l in data_tag_fuel_type:
        fuel_type.append(l.text)
        
    data_tag_distance_travelled = soup.find_all('div',class_='_2UUy0')
    for m in range(len(data_tag_distance_travelled)):
        a=data_tag_distance_travelled[m].text
        b=str(a).split('•')[2][0:7]
        Distance_travelled.append(b)
    
    data_tag_price =soup.find_all('div',class_='_2UUy0')
    for n in range(len(data_tag_price)):
        c=data_tag_distance_travelled[n].text
        d=str(c).split('•')[2][10:19]
        price.append(d)
        
    #To find the view_details urls of each vehicle we extracted to extract specification of each vehicle
    view_soup = BeautifulSoup(i.text)

    for link in view_soup.findAll('a', attrs={'href':re.compile('^/buy-used-\w+.+\d/')}):
        view_details_url_list.append(link.get('href'))
```

```python
string = 'https://www.cars24.com'
view_details_url_list = [string + x for x in view_details_url_list]
view_details_url_list
```

```python
for a in view_details_url_list:
    view_details_url_data = requests.get(str(a))
    view_details_url_data_soup = BeautifulSoup(view_details_url_data.text)
    data_tag = view_details_url_data_soup.find_all('p')
    data_tag =data_tag[29:33]
    Owner_type.append(data_tag[0].text)
    Transmission.append(data_tag[1].text)
    Rto.append(data_tag[2].text)
    Insurance.append(data_tag[3].text)
```

```python
print(len(car_names))
print(len(model_year))
print(len(fuel_type))
print(len(Distance_travelled))
print(len(price))
print(len(Owner_type))
print(len(Transmission))
print(len(Rto))
print(len(Insurance))
```

```python
used_cars_df = pd.DataFrame()
```

```python
used_cars_df['Car_Names'] = car_names
used_cars_df['Model_Year']=model_year
used_cars_df['Fuel_Type']=fuel_type
used_cars_df['Kms_Travelled']=Distance_travelled
used_cars_df['Price']=price
used_cars_df['Owner_Type']=Owner_type
used_cars_df['Transmission_Type']=Transmission
used_cars_df['RTO']=Rto
used_cars_df['Insurance_Type']=Insurance
```
